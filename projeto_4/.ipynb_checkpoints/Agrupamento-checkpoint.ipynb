```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("drug-use-by-age.csv", na_values='-').fillna(0)
df[[column for column in df.columns if "frequency" in column]]
```

```python
df_group = df[["alcohol_frequency","marijuana_frequency"]]

kmeans = KMeans(n_clusters=3)
kmeans.fit(df_group)
sns.scatterplot(data=df, x='alcohol_frequency', y='marijuana_frequency', hue=kmeans.labels_, palette='bright')
```

Aqui podemos indentificar 3 grupos, um que faz uso moderado de marijuana e baixo consumo de alcool, <br>
o segundo com alto uso de mariuaja e consumo moderado de alcool <br>
e por ultimo alto consumo de de ambas substancias

```python
df_group = df[["cocaine_frequency","crack_frequency"]]

kmeans = KMeans(n_clusters=2)
kmeans.fit(df_group)
sns.scatterplot(data=df_group, x='cocaine_frequency', y='crack_frequency', hue=kmeans.labels_, palette='bright')
```

No grafico acima podemos ver um grande grupo, onde as frequencias de cada substancia são relativamentes parecidas.
Como ambas as drogas se tratam de substâncias de efeito curto, faz sentido supor que a demanda de seus usarios é bem proxima

```python
df_group = df[["stimulant_frequency","sedative_frequency"]]

kmeans = KMeans(n_clusters=2)
kmeans.fit(df_group)
sns.scatterplot(data=df_group, x='stimulant_frequency', y='sedative_frequency', hue=kmeans.labels_, palette='bright')
```

com esse grafico podemos ver que não é comum a utilização de sedativos e estimulantes pelo mesmo usuario. 
O grupo mais a esquerda mostra que mesmo com o aumento de sedativos a frequencia de estimulantes permanece quase inalterada.
