```python
'''
Consumo de diversos tipos de drogas nos Estados Unidos em 2012
'''
import pandas as pd
import statistics as stts
import matplotlib.pyplot as plt

df = pd.read_csv("drug-use-by-age.csv")
df.head()

#as colunas com label "use" estão apresentadas por porcentagem
```

```python
# substituindo valores desconhecidos por NaN
df.replace('-', None, inplace=True)
```

```python
#alterando datatype de todas as colunas exceto as colunas 'age' e 'n' 
columns = ["age", "n"]

tochange = [col for col in df.columns if col not in columns]
df[tochange] = df[tochange].astype('float64')
```

```python
#removendo colunas com valores em porcentagem
columns = df.columns.tolist()
filter_columns = [item for item in columns if "use" not in item]

df_frequency = df[filter_columns]
df_frequency.head()
#As frequencias correspondem ao valor medio que cada usuario utilizou determinada droga em até 12 meses
```

```python
df_frequency.boxplot(column=["alcohol_frequency","marijuana_frequency","cocaine_frequency"])
plt.show()
```

```python
alcohol_median = stts.median(df_frequency["alcohol_frequency"])
marijuana_median = stts.median(df_frequency["marijuana_frequency"])

print(f"alcohol frequency Q2 = {alcohol_median}")
print(f"marijuana frequency Q2 = {marijuana_median}")
```

```python
#dataframe do uso

columns = df.columns.tolist()
filter_columns = [item for item in columns if "frequency" not in item]

df_use = df[filter_columns]
```

```python
#formatando coluna de idades

for i in range(10,16):
    pair = df_use.iloc[i,0].split('-')
    pair = [float(x) for x in pair]
    df_use.iloc[i,0] = str(stts.median(pair))
    
df_use.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].plot(df_use['age'], df_use['alcohol_use'], marker='o', color='b', label='Álcool')
axes[0, 0].set_title('Uso de Álcool por Idade')
axes[0, 0].set_xlabel('Idade')
axes[0, 0].set_ylabel('% de Uso')
axes[0, 0].legend()

axes[0, 1].plot(df_use['age'], df_use['marijuana_use'], marker='o', color='g', label='Maconha')
axes[0, 1].set_title('Uso de Maconha por Idade')
axes[0, 1].set_xlabel('Idade')
axes[0, 1].set_ylabel('% de Uso')
axes[0, 1].legend()

axes[1, 0].plot(df_use['age'], df_use['cocaine_use'], marker='o', color='r', label='Cocaína')
axes[1, 0].set_title('Uso de Cocaína por Idade')
axes[1, 0].set_xlabel('Idade')
axes[1, 0].set_ylabel('% de Uso')
axes[1, 0].legend()

axes[1, 1].plot(df_use['age'], df_use['hallucinogen_use'], marker='o', color='m', label='Alucinógenos')
axes[1, 1].set_title('Uso de Alucinogenos por Idade')
axes[1, 1].set_xlabel('Idade')
axes[1, 1].set_ylabel('% de Uso')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
#selecionando as idades com frequencia maxima de consumo por substancia
fcolumns = [item for item in df_frequency.columns.tolist() if "frequency" in item]

maximum_intake = []

for column in fcolumns:
    max_value = df_frequency[column].max()
    partition = df_frequency[df_frequency[column] == max_value][['age',column]]
    maximum_intake.append(partition)
```

```python
max_frequencies = []
substances_with_age = []

for i in range(1, len(maximum_intake)):
    age  = maximum_intake[i].iloc[:, 0].values[0]
    frqn = maximum_intake[i].iloc[:, 1].values[0]
    subs = maximum_intake[i].columns[1]

    max_frequencies.append(frqn)
    substances_with_age.append(f"{subs}\n{age}")

plt.figure(figsize=(12, 6))
plt.bar(substances_with_age, max_frequencies, color='skyblue')

plt.xlabel('Substância e Idade')
plt.ylabel('Frequência Máxima')
plt.title('Idade com Frequência Máxima de Consumo por Substância')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
age = []
frequency = []

for data in df_frequency[["age","alcohol_frequency"]].values:
    age.append(data[0])
    frequency.append(data[1])
    

plt.figure(figsize=(12, 6))
plt.plot(age, frequency, color='red')

plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Frequência do Consumo de Álcool por Idade')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
