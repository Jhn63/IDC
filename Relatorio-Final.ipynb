## Limpeza de dados

O dataframe apresenta alguns problemas da maneira que ele é lido do arquivo csv, por isso se faz necessário realizar limpezas e algumas modificações antes de trabalharmos com os dados

### Issues
* Algumas colunas do tipo frequency estão com dados faltantes, representados por "-"
* As mesmas colunas tem um dtype do tipo object, que dificultará a manipulação dos dados posteriormente
* Os dados individuais de cada participante estão ocultos, em vez disso eles estão agrupados por idade e representados como um só

```python
# ao especificar os valores "-" como NaN resolvemos o problema das colunas de dtype object, que passam a ter um dtype float
# em seguida substituimo os valores NaN com 0, para representar a ausência de dados
import pandas as pd

df = pd.read_csv("drug-use-by-age.csv", na_values='-').fillna(0)
df.head()
```

Se o data-set fornecesse o consumo individual de cada participante estaria em nossas mãos uma extensa tabela preechida por varios zeros, ja que na maioria das drogas existe uma parcela muito pequena de quem faz o uso da substância. E é essa parcela que na maior parte do tempo que vai nos interessar.

No entanto mesmo obtendo a quantidade de usuários de cada subtância, ainda faltaria saber a frequência exatada do consumo de cada um. Para facilitar o estudo iremos considerar que a frequencia de todos será igual ao valor apresentado na coluna frequency respectivo a cada droga

Segue o exemplo:

```python
n_aged12 = df.iloc[0,1]
alco_use = df.iloc[0,2]
alco_fre = df.iloc[0,3]

total_users = round(n_aged12 * alco_use/100)

print(f"Apenas {total_users} dos {n_aged12} participantes de 12 anos fizeram consumo de alcool!")
print(f"Considere que todos os {total_users} usuários consumiram alcool {alco_fre} vezes nos ultimos 12 meses")
```

# Questionamentos

### 1 - Quais são as drogas que apresentam maior frequência e o seu respectivo grupo etário?

```python
# filtrando colunas frequency do data frame
fcolumns = [item for item in df.columns if "frequency" in item] 

max_frequencies = []
drug_with_group = []

for column in fcolumns:
    idx = df[column].idxmax()              # encontrando indice do valor maximo
    age = df.iloc[idx]['age']              # selecionando idade e frequencia
    frq = df.iloc[idx][column]
    drg = column
    
    max_frequencies.append(frq)
    drug_with_group.append(f"{drg}\n{age}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(drug_with_group, max_frequencies, color='skyblue')

plt.xlabel('Substância e Idade')
plt.ylabel('Frequência Máxima')
plt.title('Idade com Frequência Máxima de Consumo por Substância')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### Resposta

De longe as drogas com maior frequência de uso são os estimulantes com 350 usos no ano, e a heroina com mais de 250; seu grupo etário são as pessoas com idade igual ou superior a 65 anos, e pessoas entre 35 e 49 anos respectivamente.

Enquanto o causa do uso excessivo de estimulantes entre pessoas com mais de 65 anos é desconhecido, as fontes apontam que esses dados são resultantes de Anfetaminas e Metilfenidatos; medicamentos tais como Ratalin

Já no caso da heroina não é dificil de adivinhar, por agir principalmente no sistema de recompensa causando fortes sensações de prazer e relaxação esse droga torna-se altamente viciante. A abstência de heroina também causa crises emocionais e fisicas, como ansiedade, fissura, espasmos musculares e calafrios. Esses fatores corroboram para que usuário precise recorrer a doses com mais frequência

### 2 - Quais faixa etárias possuem as maiores quantidade de usuários para cada droga?

```python
ucolumns = [item for item in df.columns if "use" in item]

total_users = []
drug = []
ages = []
perc = []
n = []

for column in ucolumns:
    idx = df[column].idxmax()
    age = df.iloc[idx]['age']
    grp = df.iloc[idx]['n']
    use = df.iloc[idx][column]

    n.append(grp)
    perc.append(use)
    ages.append(age)
    drug.append(column.replace("_use",""))
    total_users.append(round(grp * (use/100)))

data = {'drug': drug,
        'age' : ages,
        'users': total_users,
        'n': n,
        '%': perc}

highest_num_users = pd.DataFrame(data).sort_values(by='users', ascending=False)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3))
ax.axis('tight')
ax.axis('off')
ax.table(cellText=highest_num_users.values, colLabels=highest_num_users.columns, cellLoc='center', loc='center')

plt.title("Maior Quantidade de Usuários por Idade")
plt.show()
```

### Resposta
A grande maioria de usuários seja ele de qualquer subtância listada esta em torno da faixa de 20 anos. Dentre elas a com o maior numero de usuários é o
Álcool com 3963 usuários, cerca de 84.2% dos jovens de 22 à 23 anos. Vimos também que a frequência com que esse grupo consome alcool é cerca de 50 vezes ao ano, equivalente a uma ida no bar a cada final de semana

Obseve tambem que a maior quantidade de usuários de inalantes tem 16 anos de idade, mesmo sendo uma parcela pequena da amostra (apenas 3%). <br>A utilização de inalantes é mais comum por adolescentes pelo seu fácil acesso, muitas das drogas inalantes são preparadas de forma caseira: colas adesivas, desodorantes aerossóis e tintas solventes são os produtos mais utilizados nos preparos

### 3 - Baseado no que ja foi visto quais são as 3 drogas mas utilizadas?

```python
import numpy as np

users = (df[[item for item in df.columns if "use" in item]] / 100).mul(df['n'], axis=0).round() # obtendo apenas os usuários de cada droga
freqc = df[[item for item in df.columns if "frequency" in item]]

result = pd.DataFrame(np.multiply(users.values, freqc.values), columns=freqc.columns)
result.sum().sort_values(ascending=False)
```

```python
users.sum().sort_values(ascending=False)
```

### Resposta

* __Álcool__ - consumida 1.423.752 vezes por 32.367 usuários
* __Maconha__ - consumida 497.091 vezes por 10.040 usuários
* __Analgésicos__ - cosumida 45.280 vezes por 3.411 usuários

### 4 - Podemos afirma que os usuários de maconha também fazem uso do álcool?

```python
df["alcohol_frequency"].corr(df["marijuana_frequency"])
```

```python
from sklearn.linear_model import LinearRegression

x = df["alcohol_frequency"].values.reshape(-1, 1) 
y = df["marijuana_frequency"].values  

model = LinearRegression()
model.fit(x, y)

plt.scatter(x, y)
plt.xlabel("Álcool")
plt.ylabel("Maconha")
plt.title("Correlação Álcool e Marijuana")

y_pred = model.predict(x)
plt.plot(x, y_pred, color='red')
plt.show()
```

### Resposta
Embora as frequências de cada substância apresentem uma correlação forte positiva de 0.819, não podemos afirmar com certeza que todos que fazem o uso da maconha também consomem álcool, apenas que à um forte indicio que uma boa quantidade de individuos correspondem a esse critério.

Isso nos leva a outros questionamentos: Será que o consumo de álcool pode levar alguem a buscar o uso da maconha? ou vise-versa? ou será essas substâncias um primeiro passso para conhecer as demais listadas nesse trabalho?

### 5 - Alguma dessas drogas alcança estabilidade na frequência com que elas são consumidas?

```python
age = []
frequency = []

for data in df[["age","alcohol_frequency"]].values:
    age.append(data[0])
    frequency.append(data[1])
    

plt.figure(figsize=(12, 6))
plt.plot(age, frequency, color='red')

plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Frequência do Consumo de Álcool por Idade')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### Resposta
O álcool apesar de também sofrer quedas com a porcentagem total de usuários ao passar da idade, mantem uma boa estabilidade na frequencia com que ele é consumido apartir dos 21 anos. Por ser socialmente aceitavel, a maioria das pessoas não se sentem desconfortaveis em levar o alcoolismo para idades mais avançadas da vida. Outro fator é a baixa letalidade, estudos apontam que o tabagismo mata pelo menos 3 vezes mais que o alcool (excluido os acidentes de transito que envolvem alcoolismo). Essas caracteristicas podem esclarem o porque desse comportamento
