```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("drug-use-by-age.csv", na_values='-').fillna(0)
```

## Correlação entre Meth e Heroin frequency

```python
df_fre = df[[column for column in df.columns if "frequency" in column]]

x = df_fre["meth_frequency"].values.reshape(-1, 1) 
y = df_fre["heroin_frequency"].values  

model = LinearRegression()
model.fit(x, y)

plt.scatter(x, y)
plt.xlabel("Metanfetamina")
plt.ylabel("Heroina")
plt.title("Correlação Metanfetamina e Heroina")


y_pred = model.predict(x)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
#selecionando as idades com frequencia maxima de consumo por substancia
fcolumns = [item for item in df.columns if "frequency" in item]

maximum_intake = []

for column in fcolumns:
    max_value = df[column].max()
    partition = df[df[column] == max_value][['age',column]]
    maximum_intake.append(partition)
```

```python
max_frequencies = []
substances_with_age = []

for i in range(1, len(maximum_intake)):
    age  = maximum_intake[i].iloc[:, 0].values[0]
    frqn = maximum_intake[i].iloc[:, 1].values[0]
    subs = maximum_intake[i].columns[1]

    max_frequencies.append(frqn)
    substances_with_age.append(f"{subs}\n{age}")

plt.figure(figsize=(12, 6))
plt.bar(substances_with_age, max_frequencies, color='skyblue')

plt.xlabel('Substância e Idade')
plt.ylabel('Frequência Máxima')
plt.title('Idade com Frequência Máxima de Consumo por Substância')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
age = []
frequency = []

for data in df[["age","alcohol_frequency"]].values:
    age.append(data[0])
    frequency.append(data[1])
    

plt.figure(figsize=(12, 6))
plt.plot(age, frequency, color='red')

plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Frequência do Consumo de Álcool por Idade')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
